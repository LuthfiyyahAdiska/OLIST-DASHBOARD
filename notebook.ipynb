## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Kategori produk manakah yang menyumbang pendapatan (revenue) terbesar selama periode Semester 1 (Januari - Juni) tahun 2018, dan bagaimana perbandingan rata-rata harga produk antar kategori tersebut?

- **Pertanyaan 2:** Pada hari apa dalam seminggu (Day of Week) dan jam berapa terjadi volume transaksi tertinggi sepanjang tahun 2017-2018, guna menentukan jadwal optimal untuk koordinasi dengan mitra logistik?

### Cleaning Data

#### Fixing ... problem

```python
datetime_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for column in datetime_columns:
    orders_df[column] = pd.to_datetime(orders_df[column])

products_df.dropna(subset=['product_category_name'], inplace=True)

orders_df = orders_df.dropna(subset=['order_delivered_customer_date'])

print("Proses Cleaning Selesai!")

print(orders_df.info())

print("\nMissing values pada Products:", products_df['product_category_name'].isna().sum())
```

**Insight:** (Opsional)
- Data Integrity: Semua kolom waktu kini memiliki tipe data datetime, memungkinkan kita melakukan ekstraksi komponen waktu (bulan, hari, jam) untuk analisis tren.
- Kualitas Kategori: Dengan menghapus data produk tanpa kategori, kita menjamin bahwa hasil analisis profitabilitas per kategori produk nanti tidak akan mengandung bias "Unknown Category".
- Volume Data: Setelah dilakukan pembersihan (menghapus pesanan yang belum sampai atau kategori kosong), kita memiliki dataset yang solid dan siap tempur untuk masuk ke tahap analisis mendalam.


## Exploratory Data Analysis (EDA)

### Explore ...

```python
main_df = pd.merge(order_items_df, orders_df, on="order_id", how="inner")

main_df = pd.merge(main_df, products_df[['product_id', 'product_category_name']], on="product_id", how="inner")

main_df.sample(5)

# EDA Univariate Kategorikal
print("Top 10 Kategori Produk Berdasarkan Jumlah Pesanan:")
top_10_categories = main_df['product_category_name'].value_counts().head(10)
print(top_10_categories)
top_10_categories.plot(kind='bar', color='skyblue')

# EDA Univariate Numerikal
print("Statistik Deskriptif Harga dan Ongkos Kirim:")
print(main_df[['price', 'freight_value']].describe())

plt.figure(figsize=(10, 5))
sns.histplot(main_df[main_df['price'] <= 500]['price'], bins=50, kde=True, color='purple')
plt.title('Distribusi Harga Produk (≤ 500 BRL)')
plt.xlabel('Harga (BRL)')
plt.ylabel('Frekuensi')
plt.show()

# EDA Bivariate Numerikal
correlation = main_df[['price', 'freight_value']].corr().iloc[0, 1]
print(f"Korelasi antara Harga dan Ongkos Kirim: {correlation:.2f}")

plt.figure(figsize=(8, 6))
sns.scatterplot(data=main_df, x='price', y='freight_value', alpha=0.3)
plt.title('Hubungan antara Harga Produk dan Ongkos Kirim')
plt.xlabel('Harga Produk (BRL)')
plt.ylabel('Ongkos Kirim (BRL)')
plt.show()

# Eksplorasi Pertanyaan 1
category_performance = main_df.groupby("product_category_name").agg({
    "order_id": "nunique",
    "price": "sum"
}).rename(columns={"order_id": "order_count", "price": "revenue"}).sort_values(by="revenue", ascending=False)

print("Top 5 Kategori Berdasarkan Revenue:")
print(category_performance.head())

# Eksplorasi Pertanyaan 2
main_df['order_hour'] = main_df['order_purchase_timestamp'].dt.hour
main_df['order_day'] = main_df['order_purchase_timestamp'].dt.day_name()

hourly_orders = main_df.groupby('order_hour')['order_id'].nunique()
print("\nJam dengan transaksi tertinggi:", hourly_orders.idxmax())
```

**Insight:** (Opsional)
- Dominasi Kategori: Ditemukan bahwa kategori seperti beleza_saude (kesehatan/kecantikan) atau relogios_presentes seringkali mendominasi pendapatan, namun kategori dengan volume besar seperti bed_bath_table juga memiliki pengaruh signifikan pada logistik.
- Pola Waktu: Transaksi cenderung meningkat drastis pada jam istirahat siang (12:00 - 14:00) dan malam hari setelah jam kerja. Hari Senin dan Selasa seringkali menjadi hari dengan volume pesanan tertinggi dibandingkan akhir pekan.
- Korelasi Harga: Produk dengan harga menengah atas menyumbang revenue lebih besar meskipun secara kuantitas tidak sebanyak produk murah.

## Visualization & Explanatory Analysis

### Pertanyaan 1: 5 Kategori Produk (Revenue)

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_categories = category_performance.head(5).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(
    hue="product_category_name",
    x="revenue",
    y="product_category_name",
    data=top_categories,
    palette="viridis"
)
plt.title("Top 5 Kategori Produk Berdasarkan Total Pendapatan (H1 2018)", fontsize=14)
plt.xlabel("Total Pendapatan (BRL)")
plt.ylabel("Kategori Produk")
plt.show()
```

### Pertanyaan 2: Pola Jam Sibuk Transaksi

```python
plt.figure(figsize=(12, 6))
hourly_orders.plot(kind='line', marker='o', color='tab:blue', linewidth=2)
plt.title("Pola Transaksi Pelanggan Per Jam (2017-2018)", fontsize=14)
plt.xticks(range(0, 24))
plt.xlabel("Jam (24-Hour Format)")
plt.ylabel("Jumlah Pesanan Unik")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

daily_orders = main_df.groupby('order_day')['order_id'].nunique().reindex(days_order)

plt.figure(figsize=(10, 5))
sns.barplot(
    x=daily_orders.index,
    y=daily_orders.values,
    hue=daily_orders.index,
    palette="Blues_d",
    legend=False
)
plt.title("Total Transaksi Berdasarkan Hari (2017-2018)", fontsize=14)
plt.xlabel("Hari dalam Seminggu")
plt.ylabel("Jumlah Pesanan Unik")
plt.show()

print("Detail Jumlah Transaksi Per Hari:")
print(daily_orders)
```

**Insight:** (Opsional)
- mengidentifikasi bahwa efisiensi bisnis Olist dapat ditingkatkan dengan memfokuskan inventaris pada 5 kategori teratas dan menyelaraskan jadwal operasional logistik dengan jam sibuk pelanggan. Dengan data ini, kita tidak lagi menebak-nebak, tapi melangkah dengan angka.

## Analisis Lanjutan (Opsional)

```python
snapshot_date = main_df['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

rfm = main_df.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days,
    'order_id': 'nunique',
    'price': 'sum'
}).rename(columns={
    'order_purchase_timestamp': 'Recency',
    'order_id': 'Frequency',
    'price': 'Monetary'
})

print("RFM Analysis Summary:")
print(rfm.describe())
```

## Conclusion & Recommendation

- **Conclusion pertanyaan 1:** Kategori Health & Beauty dan Watches & Gifts merupakan kontributor pendapatan terbesar sepanjang Semester 1 2018. Hal ini menunjukkan bahwa pelanggan Olist cenderung mencari barang-barang bergaya hidup dan personal.

- **Conclusion pertanyaan 2:**
    *   Puncak Transaksi (Peak Time): Volume tertinggi terjadi pada pukul 16:00 dengan tren meningkat sejak pukul 10:00. Hal ini mengindikasikan pelanggan paling aktif berbelanja di sela-sela waktu produktif atau sebelum jam pulang kantor.
    *   Hari Terpadat & Terendah: Transaksi mencapai titik tertinggi pada hari Senin (15.498 pesanan) dan Selasa (15.287 pesanan). Sebaliknya, aktivitas terendah terjadi pada hari Sabtu (10.404 pesanan). Hal ini menunjukkan penurunan aktivitas belanja hingga ~32% di akhir pekan dibandingkan awal pekan.
    *   Alasan & Insight Logistik: Terjadinya penumpukan pesanan di awal minggu (Senin-Selasa) dan jam sore mengharuskan mitra logistik untuk meningkatkan kapasitas armada penjemputan (pickup) pada hari Rabu pagi guna memastikan SLA (Service Level Agreement) pengiriman tetap terjaga.


**Rekomendasi Action Item:**
- Optimalisasi Jadwal Penjemputan (Pickup Scheduling): Mengingat volume pesanan memuncak secara konsisten pada pukul 16:00, tetapkan jadwal utama penjemputan barang oleh mitra logistik (kurir first-mile) pada pukul 17:00 hingga 18:00. Hal ini akan memastikan mayoritas pesanan yang masuk di rentang jam kerja (10:00 - 16:00) dapat langsung diserahkan ke ekspedisi di hari yang sama (same-day dispatch).
- enyesuaian Kapasitas Tenaga Kerja (Manpower Allocation): Tingkatkan jumlah staf gudang bagian pengemasan (packer) khusus untuk shift sore pada hari Senin dan Selasa. Karena ini adalah hari dengan lonjakan pesanan terpadat, penambahan staf akan mencegah terjadinya bottleneck (penumpukan pesanan yang belum diproses) yang berisiko memperlambat waktu pengiriman.
- Target SLA Spesifik Awal Pekan: Terapkan kebijakan Service Level Agreement (SLA) internal yang lebih ketat di awal minggu. Misalnya, pastikan setiap pesanan yang masuk sebelum pukul 15:00 pada hari Senin/Selasa berstatus ready to ship dan wajib diserahkan ke pihak ekspedisi maksimal pada hari Rabu pagi.
- Distribusi Beban Kerja Akhir Pekan: Karena hari Sabtu dan Minggu memiliki volume transaksi paling rendah, manfaatkan waktu luang di akhir pekan ini untuk kegiatan operasional pemeliharaan, seperti restock barang dari supplier, stock opname (pengecekan inventaris), atau penyortiran area gudang agar siap menghadapi lonjakan pesanan di hari Senin.
